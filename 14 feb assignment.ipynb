```python
#Ans 2: threading module is used to used to manage the threading in multithreading python.
# activeCount():- This functon returns total number of threads which are active.
# currentThread():- return number of thread objects the are currently in caller's thread objects
# enumerate():-return list of objects that are currently active.
```

```python
#Ans 1: run():- entry point for the thread.
# start():- start thread by calling run method.
# join():- wait till thread terminate
# isAlive():- return true of false if thread is alive or not.
```

```python
def pwr(number,power):
    print(number**power)
```

```python
import threading

thread1=[threading.Thread(target=pwr, args=(i,2)) for i in range(10)]
thread2=[threading.Thread(target=pwr,args=(i,3)) for i in range(10)]
```

```python
for i in thread1:
    i.start()
```

```python
for i in thread2:
    i.start()
```
